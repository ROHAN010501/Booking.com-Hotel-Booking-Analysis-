## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

df['total_guests'] = df['adults'] + df['children'] + df['babies']
df.head()
````

````python
# total bookings
df_bookings = df[df['is_canceled'] == 0]
Total_guests = df_bookings.groupby(['hotel'])['total_guests'].sum().reset_index()
Total_guests
````

````python
# arrival of guests per month
guest_arrival_per_month = df.groupby('arrival_date_month')['total_guests'].sum().sort_values(ascending=False).reset_index()

guest_arrival_per_month
````

````python
# cancellation w.r.t month
df.groupby('arrival_date_month')['is_canceled'].sum().sort_values(ascending=False).reset_index()
````

````python
# checking the no. of cancellation of city and resort
df.groupby('hotel')['is_canceled'].sum().reset_index()
````

````python
# calculating total no of guests month and year wise

grouped_df = df.groupby(['arrival_date_year','arrival_date_month'])['total_guests'].sum().sort_values().reset_index()

grouped_df
````

````python
# calculating avg stay in each type

total_stays = df['stays_in_week_nights'] + df['stays_in_weekend_nights']

#creating a new column which shows total count of stays
df['total_stays'] = total_stays

df.groupby('hotel')['total_stays'].mean().reset_index()
````

````python
# top 10 countries with maximum no. of total guests

max_guests_countries = df.groupby('country')['total_guests'].sum().sort_values(ascending = False).head(10).reset_index()
guests_cnt_total = sum(df['total_guests'])
max_guests_countries['% of each country'] = round((max_guests_countries['total_guests']/guests_cnt_total)*100, 2)
max_guests_countries['% of each country cumulative'] = round((max_guests_countries['total_guests']/guests_cnt_total)*100, 2).cumsum()
max_guests_countries
````

````python
# last 10 countries with minimum no. of total guests

df.groupby('country')['total_guests'].sum().sort_values(ascending = True).head(10).reset_index()
````

````python
# type of meals preferred by the guests

df.groupby('meal')['total_guests'].sum().sort_values(ascending=False).reset_index()
````

````python
# analysing which market segment the guests belongs to

df.groupby('market_segment')['total_guests'].sum().sort_values(ascending = False).reset_index()
````

````python
# calculating which distribution channels contribute the most in booking hotels

distribution_channel = df.groupby('distribution_channel')['total_guests'].sum().sort_values(ascending = False).reset_index()
distribution_channel
````

````python
# count of previous cancellations

df['previous_cancellations'].value_counts().reset_index()
````

````python
# calclating % times guests will receive or not receive their reserved room

# creating a new column which shows reserved room received or not
df['reserved_room_received'] = df['reserved_room_type'] == df['assigned_room_type']

reserved_room_received = df.groupby(['reserved_room_received'])['total_guests'].sum().reset_index()
reserved_room_received['% chance of getting reserved room'] = round((reserved_room_received['total_guests']/reserved_room_received['total_guests'].sum())*100, 2)

reserved_room_received
````

````python
# Top 10 countries with repeated and loyal guests

loyal_guests = df[(df['is_repeated_guest']==1) & (df['is_canceled']==0) & (df['previous_cancellations'] == 0)]
loyal_guests_country = loyal_guests.groupby('country')['is_repeated_guest'].sum().sort_values(ascending=False).reset_index().iloc[:10,:]
print(loyal_guests_country)
loyal_guests_hotel = loyal_guests.groupby('hotel')['is_repeated_guest'].sum().sort_values(ascending=False).reset_index()
print(loyal_guests_hotel)
````

````python
# count of cancellations in hotels

df['cancelled'] = df['previous_cancellations'].apply(lambda x: 1 if x > 0 else 0)
df.groupby('hotel')['cancelled'].sum().reset_index()
````

````python
# Top 10 maximum no. of days in waiting list

diwl = df['days_in_waiting_list'].value_counts().reset_index()
sorted_df = diwl.sort_values(by='days_in_waiting_list', ascending=False).reset_index(drop=True)
sorted_df.head(10)
````

````python
# average no. of days the guest stay per month
df.groupby('arrival_date_month')['total_stays'].mean().sort_values(ascending=False).reset_index()
````

````python
# deposit type

df['deposit_type'].value_counts().reset_index()
````

````python
# avg waiting days required per month

grouped_df = df.groupby(['arrival_date_month', 'hotel'])['days_in_waiting_list'].mean().reset_index().sort_values('arrival_date_month',ascending = True)

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

#creating categorical variable and mainting the order
grouped_df['arrival_date_month'] = pd.Categorical(grouped_df['arrival_date_month'], categories=month_order, ordered=True)

# Sort the DataFrame by 'arrival_date_year' and the ordered 'arrival_date_month'
avg_days_in_waiting_list = grouped_df.sort_values(['arrival_date_month', 'hotel'], ascending=True).reset_index(drop = True).rename(columns={'days_in_waiting_list':'avg_days_in_waiting_list'})

avg_days_in_waiting_list

````

````python
# the no. of average guests arrival per year

year_wise_guests = df.groupby('arrival_date_year')['total_guests'].sum().reset_index().sort_values('arrival_date_year')

months_in_15 = df[df['arrival_date_year'] == 2015]['arrival_date_month'].unique()
months_in_16 = df[df['arrival_date_year'] == 2016]['arrival_date_month'].unique()
months_in_17 = df[df['arrival_date_year'] == 2017]['arrival_date_month'].unique()

# creating 2 new columns
year_wise_guests['months_data_available'] = [len(months_in_15), len(months_in_16), len(months_in_17)]
year_wise_guests['average_guests_per_year'] = year_wise_guests['total_guests']/year_wise_guests['months_data_available']

# droping the columns using drop method
year_wise_guests.drop(['total_guests','months_data_available'], axis=1, inplace=True)

year_wise_guests
````

````python
# average parking required for hotels per month

car_parking = df.groupby(['arrival_date_month','hotel'])['required_car_parking_spaces'].sum().reset_index()

car_parking['month_count'] = car_parking.groupby('arrival_date_month')['arrival_date_month'].transform('count')
car_parking['avg_car_parking_per_month'] = round(car_parking['required_car_parking_spaces']/car_parking['month_count'])
car_parking.drop(['required_car_parking_spaces','month_count'],axis=1, inplace=True)

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

#creating categorical variable and mainting the order
car_parking['arrival_date_month'] = pd.Categorical(car_parking['arrival_date_month'], categories=month_order, ordered=True)

car_parking_sorted = car_parking.sort_values(['arrival_date_month','hotel'],ascending = True)
car_parking_sorted
````

````python
#  total guests per customer type and month
df.groupby(['customer_type','arrival_date_month'])['total_guests'].sum().sort_values(ascending=False).reset_index()
````

### What all manipulations have you done and insights you found?

* **Seasonal Patterns and Guest Behavior:**

August and July emerge as peak months for hotel occupancy, attributed to their status as the end of summer. Families capitalize on warm weather for last-minute vacations and beach trips. These months not only witness the highest influx of guests but also record the longest average stays. The analysis of booking changes indicates that August and July also experience the highest number of modifications, reflecting the dynamic nature of guest plans during these peak periods.


* **Trends in Guest Arrivals:**

The analysis reveals a consistent upward trend in total guests visiting hotels each year. This trend suggests a growing preference for hotel stays, possibly influenced by economic factors, seasonal trends, or changes in travel behavior. Understanding this trend is crucial for forecasting demand and optimizing hotel operations to accommodate increasing guest numbers effectively.

* **Guest Types and Top Contributing Countries:**

Transient guests constitute the largest guest type, preferring short-term stays in hotels. This demographic insight helps hotels tailor their services and marketing strategies to cater to transient guests' preferences effectively. The top countries contributing the most guests include 'PRT', 'GBR', 'FRA', 'ESP', and 'DEU', indicating the geographic diversity of hotel guests and highlighting potential markets for targeted promotional efforts.

* **Deposit Types and Booking Preferences:**

The majority of bookings (86,251) do not require any deposit, reflecting a hotel policy favoring flexibility to attract more customers. This approach is likely aimed at reducing barriers to booking and appealing to spontaneous travelers. Conversely, a smaller number of bookings (1,038) require non-refundable deposits, typically securing bookings during peak seasons or for special promotions where hotel occupancy is at its highest. Refundable deposits (107) are the least common, suggesting a lesser preference among guests or strategic considerations by the hotel.

* **Effectiveness of Distribution Channels:**

Analyzing distribution channels reveals valuable insights into how guests book their stays. Travel agents or tour operators (TA/TO) account for the majority of bookings (143,659), underscoring their effectiveness in driving hotel reservations. Direct bookings (26,656) also play a significant role, offering advantages such as direct communication with guests and reduced commission fees. Corporate bookings (6,475) represent a notable segment, indicating strong relationships with business travelers. In contrast, bookings through Global Distribution Systems (GDS) are minimal (197), suggesting a lesser impact on hotel bookings. Undefined channels account for negligible numbers (12), indicating a need for clearer categorization or monitoring of booking sources.

* **Room Reservation Insights:**

Regarding room reservations, a substantial majority (86.07%) of guests receive the room they originally reserved, highlighting effective room management practices by the hotel. This high percentage indicates that the hotel generally meets guest expectations regarding room assignments, contributing positively to guest satisfaction. However, a notable minority (13.93%) of guests do not receive their reserved rooms, potentially due to factors such as overbooking during peak periods, maintenance issues, or guest preferences for room changes upon arrival.
